# 多重共线性的问题

求解电视（TV）、收音机（Radio）和报纸（Newspaper）广告投入上述三列特征的相关系数。  
其中相关系数，是指对于一般的矩阵 X，执行 A = corrcoef(X) 后，A 中每个值的所在行和列，反应的是原矩阵 X 中相应的列向量间的相似程度。

> corrcoef: 相关系数

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings(action='ignore')
np.set_printoptions(suppress=True) #设置输出样式  启用科学记数法
path = '.\\Advertising.csv'

# pandas读入
data = pd.read_csv(path)
df = data[['TV', 'Radio', 'Newspaper']]

#多图形展示
plt.figure(figsize=(5, 3), dpi=50)
# sns.pairplot(df, vars=["TV", "Radio", "Newspaper"],palette="husl")
sns.pairplot(df, height=3)
plt.tight_layout()
plt.show()

# 主对角线上是各个属性的直方图（分布图）
# 非对角线上是两个不同属性之间的相关图
# sns.pairplot的参考资料：https://zhuanlan.zhihu.com/p/98729226
```

## 求解上述三列特征的相关系数

对于一般的矩阵 X，执行 A = corrcoef(X) 后，A 中每个值的所在行 a 和列 b，反应的是原矩阵 X 中相应的第 a 个列向量和第 b 个列向量的
相似程度（即相关系数）


```python


cm = np.corrcoef(df.values.T)

# 控制颜色刻度即颜色深浅
sns.set(font_scale=2)

# 构造关联矩阵
cols = ['TV', 'Radio', 'Newspaper']

#hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={
#    'size': 10}, yticklabels=cols, xticklabels=cols)

hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 18}, linewidths=0.2, vmax=1.0, yticklabels=cols, xticklabels=cols)
plt.xticks(rotation=90, fontsize=16)  # 将字体进行旋转
plt.yticks(rotation=360, fontsize=16)
plt.show()
```
